```python
# Import library yang dibutuhkan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Muat dataset
file_path = '/content/house_prices.csv'
data = pd.read_csv(file_path)

# Tampilkan beberapa baris pertama data
print("Preview dataset:")
display(data.head())

# Tampilkan semua nama kolom dalam dataset
print("\nNama kolom dalam dataset:")
print(data.columns)

# Informasi umum tentang dataset
print("\nInformasi dataset:")
data.info()

# Statistik deskriptif
print("\nStatistik deskriptif dataset:")
display(data.describe())

# Pastikan kolom 'price' ada sebelum melakukan visualisasi
if 'price' in data.columns:
    # Visualisasi distribusi harga
    plt.figure(figsize=(10, 6))
    sns.histplot(data['price'], kde=True)
    plt.title("Distribusi Harga Rumah")
    plt.xlabel("Harga")
    plt.ylabel("Frekuensi")
    plt.show()
else:
    print("\nKolom 'price' tidak ditemukan dalam dataset.")

# Hanya pilih kolom numerik
numerical_data = data.select_dtypes(include=[np.number])

# Korelasi antar fitur numerik
plt.figure(figsize=(12, 8))
sns.heatmap(numerical_data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matriks Korelasi")
plt.show()


# Analisis beberapa kolom lainnya (opsional, tergantung dataset)
# Cek distribusi jumlah kamar tidur
if 'bedrooms' in data.columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(data['bedrooms'])
    plt.title("Distribusi Jumlah Kamar Tidur")
    plt.xlabel("Jumlah Kamar Tidur")
    plt.ylabel("Frekuensi")
    plt.show()
else:
    print("\nKolom 'bedrooms' tidak ditemukan dalam dataset.")
```
